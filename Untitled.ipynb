```python
def get_year(filename): #Получаем год из имени файла
    result = re.search(r'\d\d\d\d', filename)
    return (result.group(0))

folder = './babynames'
dataset = []
for yobYEAR in os.listdir(folder): #Получаем данные за каждый год
    yob_file = os.path.join(folder, yobYEAR)
    by_year = pd.read_csv(yob_file, sep=",", names =['name', 'gender', 'count'])
    by_year['year'] = get_year(yobYEAR)
    dataset.append(by_year)

dataset = pd.concat(dataset) #Объединяем все файлы в один объект
df = pd.DataFrame(dataset)

df.head()   
```

```python
#Признак year, который Pandas считал как object явно приведем к типу int
df['year'] = df['year'].astype('int')

df.info()
```

```python
def crperiods(seq, num):
    avg = len(seq) / float(num)
    out = []
    last = 0.0
    minmax = []

    while last < len(seq):
        a = []
        out.append(seq[int(last):int(last + avg)])
        a.append(1880 + int(last))
        a.append(1880 + int(last + avg -1))
        minmax.append(a)
        last += avg
        
    return minmax

crperiods(range(1880, 2010), 10)
```

```python
total = pd.DataFrame(columns=['name','gender','count','period'])
for period in crperiods(range(1880, 2010), 10):
    df_for_period = df[(df.year >= period[0]) & (df.year <= period[1])]
    M = df_for_period[df_for_period.gender == 'M'].groupby(['name','gender'])['count'].sum().sort_values(ascending=False).reset_index()[:1]
    F = df_for_period[df_for_period.gender == 'F'].groupby(['name','gender'])['count'].sum().sort_values(ascending=False).reset_index()[:1]
    M['period'] = '{0}-{1}'.format(period[0], period[1])
    F['period'] = '{0}-{1}'.format(period[0], period[1])
    total = pd.concat([total, M, F])

result = total.groupby(['period','gender', 'name'])['count'].sum()
result
```

```python
unique_names = total[['name','gender']].drop_duplicates()
for indx, person in unique_names.iterrows():
    df_plot = df[(df.name == person['name']) & (df.gender == person['gender'])].groupby(['year','name'])['count'].sum()
    df_plot.unstack().plot(figsize=(6,4)).grid(True)
    plt.title('Динамика рождаемости для ' + person['name'])
    plt.ylabel('Количество рожденных')
    plt.xlabel('Год')
    plt.xticks([1880, 1893, 1906, 1919, 1932, 1945, 1958, 1971, 1984, 1997, 2010])
```

Общий график динамики рождаемости по периодам (гистограммы)

```python
def for_binning(seq, num):            #Делим период на 10 частей
    avg = len(seq) / float(num)
    out = [min(seq)]
    last = 0.0

    while last < len(seq):
        out.append(min(seq) + int(last + avg -1))
        last += avg
        
    return out

for_binning(range(1880, 2010), 10)
```

```python
df_interval = df[:]
df_interval['period'] = pd.cut(df['year'], for_binning(range(1880, 2010), 10), include_lowest=True)
for indx, person in unique_names.iterrows():
    df_plot = df_interval[(df_interval.name == person['name']) & (df_interval.gender == person['gender'])].groupby(['period','name'])['count'].sum()
    df_plot.unstack().plot(figsize=(6,4), kind='bar')
    plt.title('Динамика рождаемости для ' + person['name'])
    plt.ylabel('Количество рожденных')
    plt.xlabel('Годы')
```

Для каждого года рассчитайте сколько имен покрывают 50% людей, визуализируйте (мы увидим разнообразие имен за каждый год)

```python
fifty_percent = {'year': [], 'count': []}
for year in range(1880, 2011):
    df_year = df[df.year == year].sort_values('count', ascending = False)
    #df_year['percentage'] = df_year['count'].value_counts(normalize = True)
    df_year['percentage'] = df_year['count']/df_year['count'].sum()  
    a = 0
    i = 0

    for percent in df_year['percentage']:
        a = a + percent
        i = i + 1
        if a >= 0.5:
            fifty_percent['year'].append(year)
            fifty_percent['count'].append(i)
            break

    
fifty_percent_df = pd.DataFrame(data=fifty_percent) 

fifty_percent_df      
```

```python
fifty_percent_df.plot(x='year', y = 'count')
```

Выберите 4 года из всего промежутка и отобразите для каждого года распределение по первой букве в имени, по последней букве в имени

```python
fourYearsPeriod = range(2000, 2004)

for year in fourYearsPeriod:
    df_four = df[df.year == year]
    df_four['first letter'] = df_four['name'].astype(str).str[0]
    df_four['last letter'] = df_four['name'].astype(str).str[-1].str.upper()
    s1 = df_four.groupby('first letter')['count'].sum()
    s2 = df_four.groupby('last letter')['count'].sum()
    s3 = pd.concat([s1, s2], axis=1)
    s3.columns = ['first', 'last']
    s3.plot(kind = 'bar')
    plt.xlabel(year)
```
